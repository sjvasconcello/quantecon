## Exercise 2
The **binomial random variable** $Y ~ Bin(n,p)$ represents the number of successes in $n$ binary trials, where each trial succeeds with probability $p$ 

```python
from numpy.random import uniform

def binomial_rv(n, p):
    count = 0
    for i in range(n):
        U = uniform()
        if U < p:
            count += 1    # Or count += 1
    return count

binomial_rv(10, 0.5)
```

```python
import numpy as np

n = 100000

count = 0
for i in range(n):
    u,v = np.random.uniform(), np.random.uniform()
    d = np.sqrt((u - 0.5)**2 + (v - 0.5)**2)
    if d < 0.5:
        count += 1
area_estimate = count / n

print(area_estimate * 4)
```

```python
import numpy as np
import matplotlib.pyplot as plt

alphas = [0.0, 0.8, 0.98]
ts_length = 200

for alpha in alphas:
    x_values = []
    current_x = 0
    for i in  range(ts_length):
        x_values.append(current_x)
        current_x = alpha * current_x + np.random.randn()
    plt.plot(x_values, label=f'alpha = {alpha}')

x = [np.random.randn() for i in range(100)]
plt.plot(x, label="white noise")
plt.legend()
plt.show()
```
